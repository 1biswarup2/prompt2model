```python
#from diffusers import StableDiffusionPipeline
import torch
# Function to generate an image from a prompt using Hugging Face's API
def generate_image(promptip):
    answers = stability_api.generate(
    prompt=promptip,
    seed=121245125, # if provided, specifying a random seed makes results deterministic
    steps=50, # defaults to 30 if not specified
    )

    # iterating over the generator produces the api response
    for resp in answers:
        for artifact in resp.artifacts:
            if artifact.finish_reason == generation.FILTER:
                warnings.warn(
                    "Your request activated the API's safety filters and could not be processed."
                    "Please modify the prompt and try again.")
            if artifact.type == generation.ARTIFACT_IMAGE:
                img = Image.open(io.BytesIO(artifact.binary))
    return img

```

```python
import logging
import os
import tempfile
import time

import gradio as gr
import numpy as np
import rembg
import torch
from PIL import Image
from functools import partial

from tsr.system import TSR
from tsr.utils import remove_background, resize_foreground, to_gradio_3d_orientation

import argparse
# Device setup for PyTorch
# device = "cuda:0" if torch.cuda.is_available() else "cpu"

# # Model initialization
# model = TSR.from_pretrained("stabilityai/TripoSR", config_name="config.yaml", weight_name="model.ckpt")
# model.renderer.set_chunk_size(8192)
# model.to(device)

# Preprocessing function

rembg_session = rembg.new_session()

def preprocess(input_image, do_remove_background, foreground_ratio):
    def fill_background(image):
        image = np.array(image).astype(np.float32) / 255.0
        image = image[:, :, :3] * image[:, :, 3:4] + (1 - image[:, :, 3:4]) * 0.5
        image = Image.fromarray((image * 255.0).astype(np.uint8))
        return image

    if do_remove_background:
        image = input_image.convert("RGB")
        image = remove_background(image, rembg_session)
        image = resize_foreground(image, foreground_ratio)
        image = fill_background(image)
    else:
        image = input_image
        if image.mode == "RGBA":
            image = fill_background(image)
    return image
# Generate 3D model function
# def generate(image, mc_resolution,formats=["obj", "glb"]):
#     scene_codes = model(image, device=device)
#     mesh = model.extract_mesh(scene_codes, resolution=mc_resolution)[0]
#     mesh = to_gradio_3d_orientation(mesh)
#     rv = []
#     for format in formats:
#         mesh_path = tempfile.NamedTemporaryFile(suffix=f".{format}", delete=False)
#         mesh.export(mesh_path.name)
#         rv.append(mesh_path.name)
#     return rv
```

```python
#Device setup for PyTorch
device = "cuda:0" if torch.cuda.is_available() else "cpu"

# Model initialization
model = TSR.from_pretrained("stabilityai/TripoSR", config_name="config.yaml", weight_name="model.ckpt")
model.renderer.set_chunk_size(8192)
model.to(device)
```

```python
import torch
import trimesh
import tempfile

def generate(image, mc_resolution):
    torch.cuda.empty_cache()  # Clear unused memory
    with torch.no_grad():  # Reduces memory usage by not saving gradients
        scene_codes = model(image, device=device)
        mesh = model.extract_mesh(scene_codes, resolution=mc_resolution)[0]
        mesh = to_gradio_3d_orientation(mesh)

    with tempfile.NamedTemporaryFile(delete=False, suffix=".obj") as tmp:
        mesh.export(tmp.name, file_type='obj')
        print(f"Saved OBJ file to {tmp.name}")
        return tmp.name

# Usage
```

```python

# Gradio interface
with gr.Blocks() as interface:
    prompt_input = gr.Textbox(label="Enter a prompt for image generation")
    do_remove_background = gr.Checkbox(label="Remove Background", value=True)
    foreground_ratio = gr.Slider(label="Foreground Ratio", minimum=0.5, maximum=1.0, value=0.85, step=0.05)
    mc_resolution = gr.Slider(label="Marching Cubes Resolution", minimum=32, maximum=320, value=256, step=32)
    submit = gr.Button("Generate")
    output_model = gr.Model3D(label="Output Model")

    submit.click(
        fn=lambda prompt: generate(preprocess(generate_image(prompt), do_remove_background.value, foreground_ratio.value), mc_resolution.value),
        inputs=prompt_input,
        outputs=output_model
    )

interface.launch()
```
